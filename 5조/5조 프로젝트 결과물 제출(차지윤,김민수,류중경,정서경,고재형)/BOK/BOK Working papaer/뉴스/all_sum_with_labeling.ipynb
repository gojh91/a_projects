```python
import pandas as pd
from ekonlpy.sentiment import MPCK
import re
```

```python
df1 = pd.read_csv('연합인포맥스n_gram.csv', header=0, index_col=0)
df2 = pd.read_csv('연합뉴스n_gram.csv', header=0, index_col=0)
df3 = pd.read_csv('edaily_token_ngram(2005_2017).csv', header=0)
```

```python
import pandas as pd
from ekonlpy.sentiment import MPCK

def text2ngram(text) :
    tokens = []
    init = ['NNG','VA','VAX','MAG','VA']
    another=['못하/VX','아니/VCN','않/VX','지만/VCP']
    for tag in text:
        if tag.split('/')[1] in init:
            tokens.append(tag)
        if tag in another:
            tokens.append(tag)
    return tokens
```

```python
for idx in range(0, len(df3.tokens)):
    try:
        b = df3.tokens[idx].split(',')
        a = text2ngram(b)
        df3.tokens[idx] = ",".join(a)
        if idx % 1000 == 0:
            print(idx)
            
    except:
        print('error{}'.format(idx))
        pass
```

```python
for idx in range(0, len(df3['date'])):
    a = df3['date'][idx]
    try:
        a = a.replace('-','.')
        df3['date'][idx] = a
        if idx%1000 == 0:
            print(len(df3['date']), idx)
    except:
        pass
```

```python
df=pd.concat([df1,df2,df3], ignore_index = True) 
```

```python
df['ngram'] = df['tokens']+df['ngram']
df = df.drop('content',axis=1).drop('title',axis=1).drop('url',axis=1).drop('tokens',axis=1)
```

```python
df = df.dropna()
df = df.sort_values('date')
df=df.reset_index(drop=True)
df
```
