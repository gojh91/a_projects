```python
import re
# tmp = re.sub(productJJanQv2,"",str(k))
JJanQ = re.sub("\/","",JJanQ)
JJanQ = re.sub("  ","",JJanQ)
JJanQ = re.sub('\|\*',"|",JJanQ)
JJanQ = re.sub('\|Q\|',"|",JJanQ)
JJanQ = re.sub("Easyshare","",JJanQ)
JJanQ = re.sub("EasyShare","",JJanQ)
JJanQ = re.sub("GPS","",JJanQ)
# JJanQ = re.sub("  ","",JJanQ)
# JJanQ = re.sub(" ","",JJanQ)
JJanQ = re.sub("\,","",JJanQ)
# JJanQ = re.sub("\| \|","|",JJanQ)
# JJanQ = re.sub("\|\|","|",JJanQ)
JJanQ = re.sub("\| ","|",JJanQ)
JJanQ = re.sub("\| ","|",JJanQ)
JJanQ = re.sub("\+ ","",JJanQ)
JJanQ = re.sub("\* ","",JJanQ)
JJanQ = re.sub(" \|","|",JJanQ)
JJanQ = re.sub("nan\|","|",JJanQ)
JJanQ = re.sub("  ","",JJanQ)
JJanQ = re.sub("\| \|","|",JJanQ)
JJanQ = re.sub("\|\|","|",JJanQ)
JJanQ = re.sub("\'","",JJanQ)
JJanQ = re.sub("\!","",JJanQ)
JJanQ = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘(\)\[\]\<\>`\'…》]', '', JJanQ)
# bran = re.findall(brandJJanQ[:len(brandJJanQ)-1],str(categ))
JJanQ
```

```python
cJJanQ = re.sub("Coolpix","CP",JJanQ)
dJJanQ = re.sub("FinePix","FP",JJanQ)
bJJanQ = re.sub(" ","",JJanQ)
JJanQ = JJanQ + "|" +cJJanQ + "|"+dJJanQ + "|"+bJJanQ 
JJanQ
```
