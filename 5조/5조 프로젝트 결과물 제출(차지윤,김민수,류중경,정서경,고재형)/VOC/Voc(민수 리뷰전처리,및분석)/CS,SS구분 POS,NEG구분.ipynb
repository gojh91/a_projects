# 데이터프레임 넣으면 pos,neg 해줌

```python
def identify_pol(df):
    import pymysql
    import pandas as pd
    import pandas as pd

    db = pymysql.connect(
        host='ec2-13-209-223-88.ap-northeast-2.compute.amazonaws.com',  # DATABASE_HOST
        port=3306,
        user='nlpuser',  # DATABASE_USERNAME
        passwd='1234',  # DATABASE_PASSWORD
        db='nlp',  # DATABASE_NAME
        charset='utf8'
    )

    SQL = "SELECT * FROM nlp.TotalDic where DicType2='MPQA';"
    df4 = pd.read_sql(SQL, db)
    
    mpqa = df4[['Rule','Label']]
    pos_mpqa = mpqa[mpqa['Label']=='POS']
    neg_mpqa = mpqa[mpqa['Label']=='NEG']
    
    df['pol'] = None
    
    for i in range(len(df)):
        pos_count = 0
        neg_count = 0
        for j in df.tagged_token[i]:
            if j[0] in pos_mpqa:
                pos_count += 1
            elif j[0] in neg_mpqa:
                neg_count += 1        
        if i%1000==0:
            print(i)
        if pos_count-neg_count > 0:
            df.pol = 'pos'
        else:
            df.pol = 'neg'
    df.to_csv('posneg생성.csv')
    return df
```
