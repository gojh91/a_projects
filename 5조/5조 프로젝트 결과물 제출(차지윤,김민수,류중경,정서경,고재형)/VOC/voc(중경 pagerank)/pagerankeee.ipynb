### Based on these sentence statistics, we build a product
graph Gf (see Fig. 2). Edge weights are determined by
comparative sentences, and node weights are determined
by subjective sentences. Since the reviews of product
C have 7 positive comparative sentences mentioning
product B (and feature F), and 2 negative comparative
sentences mentioning B (and feature f), there is an edge
from C to B with weight 3.5. It must be mentioned
that to prevent edges with infinite length (when the
number of negative comparative sentences is 0), we
set the minimum value of the denominator to 1 while

```python
from random import randint
import pandas as pd
###임시 파일 생성
Feature = ['Flash','Battery','Focus','Lens','Optical','Lcd','Resolution',
           'Burst', 'Memory','Compression']
name = ["A","B",'C',"D"]

dis = []
for idx in range(0, 100000):
    new_dict = {}
    new_dict['Feature'] = Feature[randint(0,9)]
    new_dict['Cs'] = randint(0,1)
    new_dict['Ss'] = randint(0,1)
    new_dict['name'] = name[randint(0,3)]
    new_dict['name2'] = name[randint(0,3)]
    new_dict['positive']=randint(0,10)
    new_dict['nagetive']=randint(0,10)
    dis.append(new_dict)

data = pd.DataFrame(dis)

##페이지랭크 함수

import networkx as nx
result = []
for idx in data.featureID.unique():
    G = nx.DiGraph()
    data2 = data[data.featureID == idx]
    data3 = data2[data2.Ss >= 1].copy()
    data3 = data3.reset_index(drop=True)
    data4 = data2[data2.Cs >= 1].copy()
    data4 = data4.reset_index(drop=True)

    for idx2 in range(len(data3)-1):  # for subjective sentence
        a = data3.model2[idx2]
        b = data3.model2[idx2]
        c = data3.Positive[idx2]
        d = data3.Negative[idx2]
        if d == 0:
            d = 1
        G.add_edge('{}'.format(a),'{}'.format(b), weight=c/d)

    for idx2 in range(len(data4)-1):    #for Comparetive Sentence
        a =data4.model2[idx2]
        b =data4.modelref[idx2]
        c = data4.Positive[idx2]
        d = data4.Negative[idx2]
        if d == 0:
            d = 1
        G.add_edge('{}'.format(a),'{}'.format(b), weight=c/d)

    pr = nx.pagerank(G)
    new_dict = {}
    new_dict['Feature'] = idx
    for idx in list(pr.keys()):
        new_dict['{}'.format(idx)] = pr["{}".format(idx)]
    result.append(new_dict)
    


###실행
```

    2.Cs는 (1) 번에 자신의 이름을, (2)에는 언급된 product 이름을.
    3.SS는 (1), (2)번에 자기 이름을.
    4.weight에는 positive 수/ negetive수. Negetive가 0이면 1로.
    5.결과는 dict 형태로 나옴.

```python
import pandas as pd
data = pd.read_csv('비교제품등장_페이지랭크돌리기전.csv',index_col = 0)
```

```python
data = [['featureID', 'com','pol','model','modelref',]]
```

```python
def ddddddd():
    import networkx as nx
    result = []
    for idx in data.Feature.unique():
        G = nx.DiGraph()
        data2 = data[data.Feature == idx]
        data3 = data2[data2.Ss == 1].copy()
        data3 = data3.reset_index(drop=True)
        data4 = data2[data2.Cs == 1].copy()
        data4 = data4.reset_index(drop=True)

        for idx2 in range(len(data3)-1):  # for subjective sentence
            a = data3.name[idx2]
            b = data3.name[idx2]
            c = data3.positive[idx2]
            d = data3.nagetive[idx2]
            if d == 0:
                d = 1
            G.add_edge('{}'.format(a),'{}'.format(b), weight=c/d)

        for idx2 in range(len(data4)-1):    #for Comparetive Sentence
            a =data4.name[idx2]
            b =data4.name2[idx2]
            c = data4.positive[idx2]
            d = data4.nagetive[idx2]
            if d == 0:
                d = 1
            G.add_edge('{}'.format(a),'{}'.format(b), weight=c/d)

        pr = nx.pagerank(G)
        new_dict = {}
        new_dict['Feature'] = idx
        for idx in list(pr.keys()):
            new_dict['{}'.format(idx)] = pr["{}".format(idx)]
        result.append(new_dict)
    return result
```

```python

%time ddddddd()
```
