```python
test_asin = rev[['asin','title']]
asin = test_asin.drop_duplicates(['asin'], keep='first')
asin.reset_index(drop=True)
```

```python
g = []
for i,v in enumerate(asin.title.value_counts()):
    t = {}
    t['title']=asin.title.value_counts().index[i]
    t['count'] = v
    g.append(t)
g
```

```python
asin_dict = pd.DataFrame(g)
```

```python
real_asin = asin.merge(asin_dict,how='outer',on = 'title')
```

```python
dup_asin_table = real_asin.loc[real_asin['count']>=2,:]
```

```python
dup_asin_table= dup_asin_table.reset_index(drop=True)

```

```python
fix_dup_asin_table = dup_asin_table[['title']]
fix_dup_asin_table = fix_dup_asin_table.drop_duplicates(['title'], keep='first')
fix_dup_asin_table= fix_dup_asin_table.reset_index(drop=True)
```

```python
assinlist = []

for i in range(len(fix_dup_asin_table)):
    asin = []
    for j in range(len(dup_asin_table)):
        if (dup_asin_table[['title']].loc[j][0] == fix_dup_asin_table[['title']].loc[i][0] ):
            asin.append(dup_asin_table[['asin']].loc[j][0])

    assinlist.append({'title':fix_dup_asin_table[['title']].loc[i][0],
                     'asin' :asin })
```
