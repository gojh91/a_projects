```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://finance.naver.com/research/debenture_list.nhn?&"


title_1 = []
company_1 = []
pdf_1 = []
update_1 = []
views_1 = []
# rate_list = []
for x in range(1,117,1):
    params = {
        'page': x
    }
    resp = requests.get(url,params=params)
    soup = BeautifulSoup(resp.content,'html.parser')
    #print(resp.url)
    div = soup.find('div',class_='box_type_m')
    tr = div.find_all('tr')
    
    for i in range(1,len(tr)):
        if len(tr[i]) ==11:
            title = tr[i].find('td',style='padding-left:10px').text   #제목
            company = tr[i].find_all('td')[1].text                    #증권사
            pdf = tr[i].find_all('td')[2].find('a')['href']           #pdf
            update = tr[i].find_all('td')[3].text                     #작성일
            views = tr[i].find_all('td')[4].text                      #조회수
            title_1.append(title)
            company_1.append(company)
            pdf_1.append(pdf)
            update_1.append(update)
            views_1.append(views)
            new_update = update.replace('.',"")
            #print(new_update)
            #print(company)
            #print(pdf)
            file_path = new_update + '_' + company + '_' + title   # 파일명 만들기
            #print(file_path)
            
            new_file_path = file_path.replace('?',"").replace(',',"").replace('.',"").replace('/',"").replace(':',"").replace('"',"").replace('!',"").replace('<',"").replace('>',"")
            #print(new_file_path)
            
            
            file_res = requests.get(pdf)
            
            # 파일 다운로드 
            with open("./naver_finance_report/"+new_file_path+".pdf",'wb') as f:
                f.write(file_res.content)

        else:
            pass
```

```python
import pandas as pd
analysis = pd.DataFrame()
analysis
analysis['title'] = title_1
analysis['company'] = company_1
analysis['pdf'] = pdf_1
analysis['update'] = update_1
analysis['views'] = views_1
```
