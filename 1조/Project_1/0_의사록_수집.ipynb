```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.bok.or.kr/portal/bbs/B0000245/list.do?menuNo=200761"


title_1 = []
date_1 = []
update_1 =[]
views_1 = []
pdf = []
# 총 44페이지 
for x in range(1,30,1):  
    params = {
        'pageIndex': x
    }
    resp = requests.get(url,params=params)
    soup = BeautifulSoup(resp.content,'html.parser')
    #print(resp.url)
    div = soup.find('div',class_='bdLine type2')
    li = div.find_all('li')
    
    for i in range(0,10,1):
        i=3*i
        title_date = li[i].find('div',class_='row').find('span',class_='col').text   #제목
        #print(title_date)
        update_views = li[i].find('div',class_='row').find('div',class_='col s12 dataInfo').text
        #print(update_views)
        
      
        title = title_date[:27].replace("(","").replace(")","").strip()
        date = title_date[27:].replace("(","").replace(")","").strip()

        print(title)
        print(date)
        update = update_views[:15].replace("등록일","").strip()
        #print(update)
        views = update_views[15:].replace("조회수","").strip()
        #print(views)
        title_1.append(title)
        date_1.append(date)
        update_1.append(update)
        views_1.append(views)
        
        file_path = date + '_' + title
        
        A = li[i].find('div',class_='row').find('div',class_='fileGroupSet').find('div',class_='fileGoupBox').find_all('li')[0].find('a')['title'][-3:]
        B = li[i].find('div',class_='row').find('div',class_='fileGroupSet').find('div',class_='fileGoupBox').find_all('li')[1].find('a')['title'][-3:]

        
        
        if A=='pdf':
            try:
                pdf1 = li[i].find('div',class_='row').find('div',class_='fileGroupSet').find('div',class_='fileGoupBox').find_all('li')[0].find('a')['href']
                pdf_1 = 'http://www.bok.or.kr'+pdf1
            #print(pdf_1)
                file_res = requests.get(pdf_1)
                pdf.append(pdf_1)
            except:
                pass
            
        if B=='pdf':
            try:
            
                pdf2 = li[i].find('div',class_='row').find('div',class_='fileGroupSet').find('div',class_='fileGoupBox').find_all('li')[1].find('a')['href']
                pdf_2 = 'http://www.bok.or.kr'+pdf2
                #print(pdf_2)
                file_res = requests.get(pdf_2)
                pdf.append(pdf_2)
            except:
                pass
        # 파일 다운로드 
        with open("./K_bank_minutes/"+file_path+".pdf",'wb') as f:
            f.write(file_res.content)
```
