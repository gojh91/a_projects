```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
graph = nx.DiGraph()
    
for i in range(len(products)): 
    for j in range(len(products)):
        if i==j:
            try:
                SS_val = len(df_new[(df_new['CS_SS']=='SS') & (df_new['title']==products[i]) & (df_new['p/n']=='P')])/len(df_new[(df_new['CS_SS']=='SS') & (df_new['title']==products[i]) & (df_new['p/n']=='N')])
                #weighted_edge_3[i][j]=SS_val
                graph.add_weighted_edges_from([(i,j, SS_val)])
            except:
                pass
        else:
            try :
                CS_val = len(df_new[(df_new['CS_SS']=='CS') & (df_new['title']==products[i]) & (df_new['Ref']==products[j]) & (df_new['p/n']=='P')])/len(df_new[(df_new['CS_SS']=='CS') & (df_new['title']==products[i]) & (df_new['Ref']==products[j]) & (df_new['p/n']=='N')])
                #weighted_edge_3[j][i]=CS_val
                graph.add_weighted_edges_from([(j,i, CS_val)])
            except :
                pass
```

```python
final_value = nx.pagerank(graph, nodes = '0')
```
