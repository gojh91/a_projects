```python
import numpy as np 
import nltk
```

```python
kw_dict = pd.read_csv('./keyword.txt', sep='\n', header= None)
kw_dict = pd.DataFrame(kw_dict)
kw_dict.columns = ["KW"]
kw_dict.head(10)
KW= list(kw_dict['KW'])
```

```python
B_list = []
for i in range(len(df)):
    A = nltk.word_tokenize(df['review'][i])
    B = nltk.pos_tag(A)
    B_list.append(B) 
```

```python
for idx,j in enumerate(B_list):
    #print(j)
   
    for (word,pos) in j:
        if pos in ['JJR', 'JJS', 'RBR', 'RBS']:
            df.loc[idx,'xLab'] = 'CS'
            break
        elif word in KW:
            df.loc[idx,'xLab'] = 'CS'
            break
        else:
            df.loc[idx,'xLab'] = 'SS'
```

```python
# CS로 분류된 문장중에서 product가  없으면 빼라. 
```

```python
pos_dict = pd.read_csv('positive-words.txt', sep='\n', header= None)
neg_dict = pd.read_csv('negative-words.txt', sep='\n', header= None)
pos_dict = list(pos_dict[0])
pos_dict
```

```python
neg_dict = list(neg_dict[0])
neg_dict
```

```python
df['p/n'] = ""
```

```python

for idx,i in enumerate(B_list):
    pos_count = 0
    neg_count = 0
    for (word,pos) in i:
        if word in pos_dict:
            pos_count +=1
        elif word in neg_dict:
            neg_count +=1
        
        else:
            pass
    print(pos_count, neg_count)
    if pos_count > neg_count:
        df['p/n'] = 'P'
    if pos_count == neg_count:
        df['p/n'] = 'nothing'
    if pos_count < neg_count:
        df['p/n'] = 'N'

        
        
```
