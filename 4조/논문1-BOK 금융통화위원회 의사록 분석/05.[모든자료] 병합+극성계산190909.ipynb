# 3. up, down, neutrality df 만들어서 갯수 비교. 3개의 열 합이 total과 같은지

```python
def sep_df(temp_df):

    total_df= pd.DataFrame(temp_df['ngram'].value_counts()).reset_index() # value_counts사용
    up_df = pd.DataFrame(temp_df[temp_df.updown=='up']['ngram'].value_counts()).reset_index()
    down_df = pd.DataFrame(temp_df[temp_df.updown=='down']['ngram'].value_counts()).reset_index()
    ne_df = pd.DataFrame(temp_df[temp_df.updown=='neutrality']['ngram'].value_counts()).reset_index()
    
    
    total_df = total_df.rename(columns={'ngram':'total_count'})
    up_df = up_df.rename(columns={'ngram':'up_count'})
    down_df = down_df.rename(columns={'ngram':'down_count'})
    ne_df = ne_df.rename(columns={'ngram':'ne_count'})
    
    
    return total_df, up_df, down_df, ne_df
```

```python
total_df, up_df, down_df, ne_df = sep_df(temp_df)
```

# 5. polarity score 구하기 - up down df만 가지고 계산해서 만듦

#### (하나의 N-gram의 Pos 개수 / 전체 Pos 개수) / (하나의 N-gram의 Neg 개수 / 전체 Neg 개수)

```python
def polar_score(up_df, down_df):
    all_pos = up_df['up_count'].sum()
    all_neg = down_df['down_count'].sum()
    print(all_pos, all_neg)
    
    up_df['분자']=up_df['up_count']/all_pos
    down_df['분모']=down_df['down_count']/all_neg
    
    polar = pd.merge(up_df,down_df, how='outer', on='index')
    polar['polarity'] = polar['분자'] / polar['분모']
    polar = polar.dropna().reset_index(drop=True)
    
    return polar
```

# 7. hawk dovish 라벨링

0.75 보다 작으면 dovish

1.3보다 크면 hawkish

```python
def polar_label(polar):
    
    polar['hawk/dov']= 'hawkish'
    polar['hawk/dov']=polar[polar['polarity']<0.76]['hawk/dov'].replace('hawkish','dovish')
    polar['hawk/dov']=polar['hawk/dov'].fillna('hawkish')

    return polar
```
