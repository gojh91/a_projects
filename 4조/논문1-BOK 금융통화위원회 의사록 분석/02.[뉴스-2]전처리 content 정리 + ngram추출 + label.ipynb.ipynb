```python
import pandas as pd
import datetime
import re
import os

def check_file(path_dir):
    os.chdir(path_dir) # 작업경로 바꾸기
    
    file_list = os.listdir(path_dir) # 현재 경로 파일목록 조회

    file_list.sort()

    return file_list




# 파일 목록 중 내가 작업하기 원하는 파일만 골라내어 target_list로 만든다


def make_target_list(file_list):
    
    # 파일 이름에 특정 string을 가지는 파일만 뽑아내기
    target_list = []

    for item  in file_list: 

        # '.csv이 파일 이름에 없을 경우 -1을 반환'
        if item.find('통합뉴스') is not -1:
            target_list.append(item)
    
    print(target_list)
    
    return target_list


# 원하는 파일을 append한다

def append_all(target_list):
    
    c_list = list(pd.read_csv(target_list[0], encoding='utf-8').columns)
    total_df = pd.DataFrame(columns=c_list, index=range(0))

    for item in target_list:
        temp_df = pd.read_csv(item, encoding='utf-8')
        total_df = total_df.append(temp_df)
    
    total_df = total_df.reset_index(drop=True)
    
    return total_df



# 묶어서 한 번에 돌리기

def run_append(path_dir):
    file_list = check_file(path_dir)
    target_list = make_target_list(file_list)
    total_df = append_all(target_list)
    
    return total_df
```

```python
def clear(news, content):

    news[content] = news[content].astype('str')

    # 영어, n,t,nn,nt,@, #,$,등등 불필요한 문자 삭제
    news[content] = news[content].apply(lambda x: re.sub("[\\n\\t\[\]\\\\ntnn\r'끝)(@#^*]",'',x))

    # (서울or뉴욕=연합인포맥스or연합뉴스) 지우기
    news[content] = news[content].apply(lambda x: re.sub("[가-힣].=연합인포맥스|[가-힣].=연합뉴스|[가-힣].=이데일리|이데일리|연합뉴스|연합인포맥스",'',x))

    # 기자=이바보 포함 앞 부분 다 지우기 
#     news[content] = news[content].apply(lambda x: re.sub(".+['기자']{2}\=|.+['기자']{2} \=|.+['특파원']{2}\=|.+['특파원']{2} \=",'',x,2))
    news[content] = news[content].apply(lambda x: re.sub("[특파원,기자]",'',x))
    
    
    return news
```

## 2-1 추출하기

### return 값을 2개 반환하여 ngram과 tokens를 동시에 받아온다

```python

import pandas as pd
from ekonlpy.sentiment import MPCK


# return을 ngram과 token 각각 받는다
def text2ngram(text):
    mpck = MPCK()
    bef_tokens = mpck.tokenize(text)
    ngrams = mpck.ngramize(bef_tokens)
    
    return ngrams, bef_tokens 
```
