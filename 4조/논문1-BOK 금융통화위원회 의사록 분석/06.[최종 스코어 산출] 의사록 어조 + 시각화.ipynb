## 2-1 hawk, dovish list만들기

매파 ngram 사전 숫자 13767개  -> 15456개 -> 16124개

비둘기파 ngram 사전 숫자 12576개  -> 13977개 -> 14565개

```python
def hawk_dov_list(our_polar):
    hawk = our_polar[our_polar['hawk/dov'] == 'hawkish']
    hawk_list = hawk['ngram']
    hawk_list = list(hawk_list)
    
    dov = our_polar[our_polar['hawk/dov'] == 'dovish']
    dov = dov.reset_index(drop=True)
    dov_list = dov['ngram']
    dov_list = list(dov_list)
    
    return hawk_list, dov_list
```

# 3. 문장별 매파/비둘기파 카운트

## ngram이 hawk, dov 리스트에 속할 경우 카운트+1

```python
def dov_hawk_count(dr_rock):
    

    dr_rock['sentence_score'] = 0
    dr_rock['sentence_score'] = dr_rock['sentence_score'].astype('float') 
    dr_rock['no_h_tone'] = 0
    dr_rock['no_d_tone'] = 0
    dr_rock['h_feature'] = None
    dr_rock['d_feature'] = None


    # test_dr

    for idx,val in enumerate(dr_rock['ngram']):
        word_list = val.replace(' ','').split(',')
        h_count = 0
        d_count = 0
        sentence_score = 0
        h_feature = []
        d_feature = []



        for i in word_list:
            if i in hawk_list:
                h_count += 1
                h_feature.append(i)

            elif i in dov_list:
                d_count += 1
                d_feature.append(i)

            else:
                continue

        bunja = float(h_count - d_count)
        bunmo = float(h_count + d_count)




        try:
            sentence_score = float(bunja / bunmo)
            dr_rock['sentence_score'][idx] = sentence_score
            dr_rock['no_h_tone'][idx] = h_count
            dr_rock['no_d_tone'][idx] = d_count
            dr_rock['h_feature'][idx] = h_feature
            dr_rock['d_feature'][idx] = d_feature

        except ZeroDivisionError:
            print("ZeroDivision")
            dr_rock['sentence_score'][idx] = 0
            dr_rock['no_h_tone'][idx] = h_count
            dr_rock['no_d_tone'][idx] = d_count
            dr_rock['h_feature'][idx] = h_feature
            dr_rock['d_feature'][idx] = d_feature

        if idx % 100 == 0 :
            print('현재 인덱스 : ',idx,'카운트 : ', sentence_score)
    
    return dr_rock
            

# 출처: https://korbillgates.tistory.com/104 [생물정보학자의 블로그]
```

## 의사록 1건마다 스코어 산출

```python
def doc_score(dr_save):
    

    dr_score = pd.DataFrame(columns=['date', 'score'])
    dr_score['score'] = dr_score['score'].astype('float')

    for i in date_list:
        one_dr = dr_save[dr_save['date'] == i]
        pos_sentence = len(one_dr[one_dr['sentence_score'] > 0])
        neg_sentence = len(one_dr[one_dr['sentence_score'] < 0])


        bunmo = pos_sentence + neg_sentence
        bunja = pos_sentence - neg_sentence

        try:
            score = float(bunja / bunmo)
            dr_score = dr_score.append(pd.DataFrame([[i, score]], columns=['date','score']), ignore_index = True)

        except ZeroDivisionError:
            print("ZeroDivision")
            dr_score = dr_score.append(pd.DataFrame([[i, 0]], columns=['date','score']), ignore_index = True)

    
    return dr_score
    
    
    
## 참조 사이트
# https://shydev.tistory.com/29
```

## 7-2 정규화? 표준화? 하는 작업 1

#### 의사록score와 기준금리bmr 각각을 0~1 사이의 값으로 바꿔준다

```python
from sklearn import preprocessing

def standard(new_rate, dr_score):
    new_rate['BMR']=preprocessing.MinMaxScaler().fit_transform(new_rate['BMR'].values.reshape(-1,1))
    dr_score['score']=preprocessing.MinMaxScaler().fit_transform(dr_score['score'].values.reshape(-1,1))
    
    new_rate['date'] = pd.to_datetime(new_rate['date'])
    new_rate = new_rate.set_index('date')
    
    # -1 ~ 1 사이의 값으로 바꿈
    new_rate['BMR'] = new_rate['BMR']-0.5
    new_rate['BMR'] = new_rate['BMR'] *2
    
    dr_score['score'] = dr_score['score'] - 0.5
    dr_score['score'] = dr_score['score'] *2
    
    return new_rate, dr_score
```

```python
new_rate, dr_score = standard(new_rate, dr_score)
```
