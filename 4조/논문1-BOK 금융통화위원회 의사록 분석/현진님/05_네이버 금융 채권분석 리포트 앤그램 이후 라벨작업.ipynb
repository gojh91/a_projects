# 파일 불러와서 컬럼 순서 변경하고 df에 저장

```python
import pandas as pd
new_df = pd.read_csv("04naver_finance_report_sum.csv",encoding = 'utf-8')
```

```python
new_df.head(1)
```

```python
new_df.drop(['Unnamed: 0'], axis='columns', inplace = True)
```

```python
df = new_df
```

```python
df.head(1)
```

```python
df.drop(['Unnamed: 0.1'], axis='columns', inplace = True)
```

```python
df.head(2)
```

```python
df = df[['date','company','title','content','tokens_ngram']]
df.head(1)
```

```python
call_rate = pd.read_csv("callrate_final.csv",encoding = 'utf-8')
call_rate.head(1)
```

```python
import pandas as pd 
A = pd.merge(df,call_rate,on='date',how='left')
A.head(1)
```

```python
del A['old_date']
del A['callrate']
del A['old_callrate']
A.head(1)
```

```python
A = A.fillna(method = 'ffill')
if A.isnull().values.any():
    A = A.fillna(method = 'bfill')
```

# 네이버 금융 채권분석 리포트 라벨작업 저장

```python
A.to_csv("05naver_finance_report_label.csv",encoding = 'utf-8')
```
