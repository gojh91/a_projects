```python
#데이트 칼럼에 떵값들어가있어서 삭제
df=df[df['date'] !='date..']
```

```python
#데이트 칼럼 타입을 데이트타임으로 바꿈. 병합 위해서
df.loc[:,'date']=pd.to_datetime(df.loc[:,'date'],format='%Y-%m-%d')
```

```python
df=df.sort_values(['date'],ascending=[True])
df=df.reset_index(drop=True)
```

```python
#병ㅋ합ㅋ
token_call = pd.merge(df, call_df,on='date', how='left')
token_call
```

```python
#병ㅋ합ㅋ
token_call = pd.merge(df, call_df,on='date', how='left')
token_call
```

```python
#필요한 칼럼만 뽑아서 다시 데이따프레임 만듦
token_call=token_call.loc[:,['date','token', 'up_down']]
# token_call=token_call.loc[:,['date','ngrams', 'up_down']]
```

```python
#중간 값 다 날림
token_call=token_call[token_call['up_down'] != 0]
```

```python
#병합 안된거 검증
token_call=token_call.fillna(1)
token_call=token_call[token_call['token'] != 1]
```

```python
#csv 만듦
token_call.to_csv('news_token_labeling.csv',index=False, encoding='utf-8')
# token_call.to_csv('news_ngram_labeling.csv',index=False, encoding='utf-8')
```
