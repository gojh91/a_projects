```python
up_ngram = []
ngrams = pd.DataFrame()

for ngram in added_df[added_df['up_down'] == '상승']['ngrams']:
    split = ngram.split(',')
    count = 0
    break_ = 0
    for count in range(len(split)):
        up_ngram.append(split[count])
    break_ += 1
    if break_ % 100 == 0:
        print(break_)
```

```python
down_ngram = []


for ngram in added_df[added_df['up_down'] == '하락']['ngrams']:
    split = ngram.split(',')
    count = 0
    break_ = 0
    for count in range(len(split)):
        down_ngram.append(split[count])
    break_ += 1
    if break_ % 100 == 0:
        print(break_)
```

# news token 불용어 처리

```python

from nltk.corpus import stopwords


stop_words = ['NR',
 'VX', 'VCP', 'VCN', 'MM', 'MAJ', 'IC',
 'JKS', 'JKC', 'JKG', 'JKO', 'JKB',
 'JKV', 'JKQ', 'JX', 'JC', 'EP', 'EF',
 'EC', 'ETN', 'ETM', 'XPN', 'XSN',
 'XSV', 'XSA', 'XR', 'SF', 'SE',
 'SSO', 'SSC', 'SC', 'SY', 'SL',
 'SH', 'SN']


up_result = []
down_result = []


for token_group in added_df['token']:
    token = token_group.split(',')
    

# 상승 토큰
for token in added_df[added_df['up_down'] == '상승']['token']:
    split = token.split(',')
    count = 0
    break_ = 0
    for t in split:
        if len(t.split('/')) > 1:
            if t.split('/')[1] not in stop_words:
                up_result.append(t)

# 하락 토큰
for token in added_df[added_df['up_down'] == '하락']['token']:
    split = token.split(',')
    count = 0
    break_ = 0
    for t in split:
        if len(t.split('/')) > 1:
            if t.split('/')[1] not in stop_words:
                down_result.append(t)

```

```python

up_ngram_test = up_ngram
down_ngram_test =down_ngram
```

# up_gram, down_gram에 각각 extend 후 각각 데이터 프레임에 저장

```python
up_ngram_test.extend(up_result)
down_ngram_test.extend(down_result)
```
