```python
minute_df=pd.read_csv('minutes_wordprocessing.csv', engine='python', encoding='utf-8')
```

```python
# news_df=pd.read_csv('news_wordprocessing.csv', engine='python', encoding='utf-8')
news_ngram=pd.read_csv('news_ngram_labeling.csv', engine='python', encoding='utf-8')
news_token=pd.read_csv('token_wordprocessing.csv', engine='python', encoding='utf-8')
```

```python
# news_ngram
```

```python
# news_token
```

```python
print(len(news_ngram), len(news_token))
```

```python
news_ngram=news_ngram.rename({'ngrams':'n-gram', 'up_down':'label'},axis=1)
```

```python
news_token=news_token.rename({'token':'n-gram', 'up_down':'label'},axis=1)
```

```python
# news_ngram
```

```python
news_ngram['n-gram']=news_ngram['n-gram'].apply(lambda x:x.replace('[',"").replace(']',"").replace('[',"").replace("'",""))
```

```python
news_ngram['n-gram']=news_ngram['n-gram'].apply(lambda x:x.replace(', ',','))
```

```python
# tot_df=pd.concat([bond_df,minute_df,news_df])
tot_df=pd.concat([bond_df,minute_df,news_ngram, news_token],sort=False)
```

```python
# tot_df
```

```python
tot_df=tot_df.fillna(1)
tot_df=tot_df[tot_df['n-gram']!=1]
```

```python
tot_df['n-gram'] = tot_df['n-gram'].apply(lambda x:x.replace(', ',','))
```
