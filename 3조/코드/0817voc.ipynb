```python
import re
import time
import requests
from scrapy.http import TextResponse

def getProductName(asinList):
    returnList = []
    errorList = []
    headers = {
        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
        'accept-encoding': 'gzip, deflate, br',
        'accept-language': 'en-US,en;q=0.9,ko;q=0.8',
        'user-agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Whale/1.5.75.9 Safari/537.36',
    }
    
    start = time.time()
    for asin in asinList:
        URL = "https://www.amazon.com/product-reviews/{}".format(asin)
        resp = requests.get(URL, headers = headers)

        response = TextResponse(resp.url, body=resp.text, encoding="utf-8")
        link = response.css('body link[rel="canonical"]::attr(href)').get()
        try:
            productName = re.sub(r"(https:\/\/.*\.com\/)|(product-reviews\/)|(\?.*)","",link).split("/")[0]
            returnList.append((asin, productName))
            if(len(returnList)%5==0):
                print("{}개".format(len(returnList)))
        except:
            errorList.append(asin)
    print("time: ", time.time() - start)
    
    print("에러: {}".format(errorList))
    return returnList
```

## cs는 new_cs == 'cs' & len(nodes) > 1 인 행을 추출해서 만듦
#### cs_df도 마찬가지

```python
def func1(inputDf):
    featL = list(inputDf[ (inputDf['new_cs']=='cs') & (inputDf['nodes']!='') ]['feature'])
    sentL = list(inputDf[ (inputDf['new_cs']=='cs') & (inputDf['nodes']!='') ]['sentence'])
    nodeL = list(inputDf[ (inputDf['new_cs']=='cs') & (inputDf['nodes']!='') ]['nodes'])
    baseL = list(inputDf[ (inputDf['new_cs']=='cs') & (inputDf['nodes']!='') ]['name'])
    posnegL = list(inputDf[ (inputDf['new_cs']=='cs') & (inputDf['nodes']!='') ]['sentiment'])
    resultList = []
    for i in range(len(sentL)):
        newNodes = list(set([word.strip() for word in nodeL[i].split(",") if word.strip() != '']))
        base = baseL[i]
        sent = sentL[i]
        feat = featL[i]
        posnegTag = posnegL[i]
        if base == None:
            continue
        if base.lower() in newNodes:
            newNodes.remove(base.lower())

        # 2개 이상
        if len(list(set(newNodes))) > 1:
            if (base in newNodes):
                newNodes.remove(base)
            rl = list(map(lambda x: (feat, sent, "cs", posnegTag, base, x), newNodes))
            resultList.extend(rl)

        # 1개 
        else:
            if len(newNodes) != 0 and  base.lower() not in newNodes:
                if 'this' in sent.lower():
                    resultList.append( (feat, sent, "cs", posnegTag, base, newNodes[0]) )
            else:
                continue

    outputDf = pd.DataFrame()
    outputDf['Feature'], outputDf['Sentence'], outputDf['SentenceType'], outputDf['Sentiment'], outputDf['BaseProduct'], outputDf['RefProduct'] = \
        [ i[0] for i in resultList ], [ i[1] for i in resultList ], [ i[2] for i in resultList ], [ i[3] for i in resultList ], [ i[4] for i in resultList ], [ i[5] for i in resultList ]
    return outputDf
```

# pRank돌려보기

```python
def make_matrix_get_score(df): 
    unique_product = sorted(list(set(df['BaseProduct'])))
    node_num = len(set(df['BaseProduct']))
    import numpy as np
    matrix = np.zeros([node_num,node_num])
    print('영행렬 생성완료')
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if i==j:
                bunja = len(df[(df['SentenceType']=='ss') & (df['BaseProduct']==unique_product[i]) & (df['Sentiment']=='pos')])
                bunmo = len(df[(df['SentenceType']=='ss') & (df['BaseProduct']==unique_product[i]) & (df['Sentiment']=='neg')])
                if bunmo == 0:
                    bunmo = 1
                matrix[i,j] = bunja / bunmo
            else:
                bunja = len(df[(df['SentenceType']=='cs') & (df['BaseProduct']==unique_product[j]) & (df['RefProduct']==unique_product[i]) &(df['Sentiment']=='pos')])
                bunmo = len(df[(df['SentenceType']=='cs') & (df['BaseProduct']==unique_product[j]) & (df['RefProduct']==unique_product[i]) &(df['Sentiment']=='neg')])
                if bunja==0 and bunmo==0:
                    matrix[i,j] = 0
                    continue
                if bunja == 0:
                    bunja = 1
                if bunmo == 0:
                    bunmo = 1
                matrix[i,j] = bunja / bunmo
        print(np.round(i/len(matrix)*100,2), '% 완료')
    print('매트릭스 생성완료')
    import networkx as nx
    G = nx.DiGraph()
    for i in range(len(unique_product)):
        for j in range(len(unique_product)):
            G.add_edge(unique_product[i],unique_product[j],weight=matrix[i,j])
    print('엣지가중치 입력완료')
    score=nx.pagerank(G)
    return score
```
